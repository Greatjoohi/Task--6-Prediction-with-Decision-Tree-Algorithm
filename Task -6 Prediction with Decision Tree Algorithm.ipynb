```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
dir(load_iris())
```

```python
data=load_iris()
```

```python
data.target_names
```

```python
df=pd.DataFrame(data.data , columns=data.feature_names)
```

```python
df
```

```python
df['target']=data.target
df['species']=df.target.apply(lambda x:data.target_names[x])
```

```python
df
```

# data Visualization

```python
sns.pairplot(df,hue="species",vars=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'])
```

```python
df=df.drop('species',axis='columns')
```

# plotting correlation witth heat map

# preparing the model

```python
X=df.drop("target",axis="columns")
```

```python
X.head()
```

```python
y=df["target"]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=200)
```

# model Buliding

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X,y)
```

```python
dtree.score(X_test,y_test)
```

# Visualization of Decision Tree

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(18,18))
plot=plot_tree(dtree,
              feature_names=data.feature_names,
              class_names=data.target_names,
              filled=True,
              rounded=True,
              fontsize=14)
```
